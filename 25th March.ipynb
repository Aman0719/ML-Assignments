#### Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?

#### Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

```python
#Ans-

import pandas as pd
import matplotlib.pyplot as plt

# Create a histogram of the flight prices
plt.hist(df['Price'],edgecolor='black', bins=50)
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.show()
```

### Conclusion:

##### The histogram shows that the distribution of flight prices is skewed to the right, with a long tail of higher-priced flights.

#### Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

```python
#Ans-

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size and create a boxplot of flight prices by airline
plt.figure(figsize=(20, 6))  # adjust figure size here
sns.boxplot(x='Airline', y='Price', data=df)
plt.xlabel('Airline')
plt.ylabel('Flight Price')
plt.title('Flight Prices by Airline')
plt.show()

```

```python
The resulting boxplot will show the variation in flight prices by airline. 
Each box represents the interquartile range (IQR) of flight prices for a particular airline, with the horizontal line inside the box indicating the median price. 
The whiskers extend to the most extreme data points within 1.5 times the IQR from the edge of the box, and any points outside this range are plotted as individual points.

The boxplot shows that flight prices vary widely by airline, with some airlines having higher median prices and more variability than others. 
For example, Jet Airways and Air India have higher median prices and a wider range of prices compared to other airlines. 
On the other hand, IndiGo and SpiceJet have lower median prices and a narrower range of prices.
```

#### Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

#### Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
#Ans-

data=data[data['Type']!='0']

data.Type.value_counts()
```

```python
chars_to_remove = ['+',',','$']
cols_to_clean = ['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        data[cols]=data[cols].str.replace(item,'')
```

```python
data['Price']=data['Price'].astype(float)
```

```python
avg_price_by_type = data.groupby('Type')['Price'].mean().reset_index()

# Create a bar chart using Seaborn
sns.barplot(x='Price', y='Type', data=avg_price_by_type)
```

##### From the above graph,  We can clearly see that the free apps has no price, and the paid apps has an average price of 14$
